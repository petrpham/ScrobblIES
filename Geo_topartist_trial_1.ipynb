```python
from bs4 import BeautifulSoup
import requests
from lxml import html
import pandas as pd
```

```python
states = pd.read_csv("countries_cut_one.csv", sep =";", encoding='latin-1')
rqst_artist = 'https://ws.audioscrobbler.com/2.0/?method=geo.gettopartists&country=' + states + '&api_key=a96a3d9d79f3af63343555c8ba1451f6'
print(rqst_artist.iloc[9]['COUNTRY'])
```

```python
class Downloader():
    """
    A class designed to scrape the data based on a provided link
    """
    def __init__(self, name):
        """
        Initialization
        """
        print('Initialized downloader')
        self.name = name
        
        
                 
    def processTheLink(self,url):
        """
        Processing a given url
        """
        #self.
        self.url = url
        self.page = requests.get(url)
        self.soup = BeautifulSoup(self.page.text,'lxml')
        
```
